# Erstellen einer Adjazenzmatrix

```python
def leftshift(ba, count):
    return ba[count:] + (bitarray('0') * count)

def rightshift(ba, count):
    return (bitarray('0') * count) + ba[:-count]

```

```python
def create_adjacency(nrows, ncols):
    A = np.zeros((nrows,ncols))
    
    configs = []
    for y in range(4096):
        num = bin(y)[2:]
        num = ("0"*(12-len(num)))+num
        configs.append(bitarray(num))



    for config in configs:
        row = A[ba2int(config)]
        mask = bitarray("000000000001")
        for i in range(11):
            row[ba2int(config | mask)] = 1
            mask = leftshift(mask, 1)    

    return A
```

### Breitensuche als Graphsuche

```python
def bfs_gs(A, start_state, goal_state):
    nodes = len(A)

    # Liste besuchter Knoten
    V = []

    # Liste noch zu besuchender Knoten
    L = []

    L.append(start_state)

    while True:

        node = L[0]
        V.append(node)

        if node == goal_state: break

        L.pop(0)
        # print(f"At node {node}")
        
        i=0
        for nextnode in A[node]:
            if nextnode == 1.0:
                if not i in V + L:
                    L.append(i)
            i+=1

        # print(f"new L: {L}")
        
        
       
    
    return V, L
```

```python

V, L = bfs_gs(A, start_state, goal_state)
print(V)
```

### Tiefensuche als Graphsuche

```python
def dfs_gs(A, start_state, goal_state):
    nodes = len(A)

    # Liste besuchter Knoten
    V = []

    # Liste noch zu besuchender Knoten
    L = []

    L.append(start_state)

    while True:

        node = L[0]
        V.append(node)

        if node == goal_state: break

        L.pop(0)
        # print(f"At node {node}")
        
        i=0
        for nextnode in A[node]:
            children = []
            if nextnode == 1.0:
                if not i in V + L:
                    children.append(i)
            L = children + L
            i+=1

        # print(f"new L: {L}")
        
        
       
    
    return V, L
```

```python
V, L = dfs_gs(A, start_state, goal_state)
print(V)
```
